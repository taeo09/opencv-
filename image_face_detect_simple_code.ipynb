```python
# =======================================
# üì∏ OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄ Ïù∏ÏãùÌïòÍ∏∞ (Google Colab)
# =======================================

# 1Ô∏è‚É£ OpenCV ÏÑ§Ïπò
!pip install opencv-python

# 2Ô∏è‚É£ ColabÏóê ÌååÏùº ÏóÖÎ°úÎìú
# - Ïù¥ ÏΩîÎìú Ïã§Ìñâ ÌõÑ ÌååÏùº ÏóÖÎ°úÎìú Ï∞ΩÏù¥ Îú®Î©¥
#   Î∞îÌÉïÌôîÎ©¥Ïóê Ï†ÄÏû•Ìïú Ïù¥ÎØ∏ÏßÄÎ•º ÏÑ†ÌÉùÌïòÏÑ∏Ïöî.
from google.colab import files
uploaded = files.upload()

# ÏóÖÎ°úÎìúÌïú ÌååÏùº Ïù¥Î¶Ñ ÌôïÏù∏
import os
for filename in uploaded.keys():
    print(f"‚úÖ ÏóÖÎ°úÎìú ÏôÑÎ£å: {filename}")
    image_path = filename  # ÌååÏùº Í≤ΩÎ°ú Ï†ÄÏû•

# 3Ô∏è‚É£ OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞
import cv2
import matplotlib.pyplot as plt

# Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞ (cv2.imreadÎäî Í∏∞Î≥∏Ï†ÅÏúºÎ°ú BGR ÌòïÏãù)
img = cv2.imread(image_path)

# BGR ‚Üí RGB Î≥ÄÌôò (matplotlibÏùÄ RGB ÏÇ¨Ïö©)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•
plt.imshow(img_rgb)
plt.axis('off')
plt.title("ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ")
plt.show()

# 4Ô∏è‚É£ ÏñºÍµ¥ Ïù∏Ïãù (ÏòàÏãú)
# - OpenCVÏóêÏÑú Ï†úÍ≥µÌïòÎäî Í∏∞Î≥∏ Haar Cascade ÏÇ¨Ïö©
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Í∑∏Î†àÏù¥Ïä§ÏºÄÏùºÎ°ú Î≥ÄÌôò (ÏñºÍµ¥ ÌÉêÏßÄ Ï†ïÌôïÎèÑ Ìñ•ÏÉÅ)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# ÏñºÍµ¥ Í≤ÄÏ∂ú
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Í≤ÄÏ∂úÎêú ÏñºÍµ¥ Ï£ºÏúÑÏóê ÏÇ¨Í∞ÅÌòï Í∑∏Î¶¨Í∏∞
for (x, y, w, h) in faces:
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Í≤∞Í≥º Ï∂úÎ†•
plt.imshow(img_rgb)
plt.axis('off')
plt.title(f"Ïù∏ÏãùÎêú ÏñºÍµ¥ Ïàò: {len(faces)}")
plt.show()

# 5Ô∏è‚É£ Í≤∞Í≥º Ï†ÄÏû• (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
# Colab ÌôòÍ≤ΩÏóê Ïù∏Ïãù Í≤∞Í≥º Ï†ÄÏû• ÌõÑ Îã§Ïö¥Î°úÎìú Í∞ÄÎä•
output_path = "result.jpg"
cv2.imwrite(output_path, cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))
files.download(output_path)
```

```python
# üì∏ OpenCVÎ°ú Î∞îÌÉïÌôîÎ©¥Ïùò ÏÇ¨ÏßÑ Î∂àÎü¨Ïò§Í∏∞ & Ïù∏ÏãùÌïòÍ∏∞
# -------------------------------------------------

# 1Ô∏è‚É£ ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò Î∞è Î∂àÎü¨Ïò§Í∏∞
!pip install opencv-python-headless
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np

# 2Ô∏è‚É£ Î∞îÌÉïÌôîÎ©¥ ÏÇ¨ÏßÑ ÏóÖÎ°úÎìú
print("üìÇ Î∞îÌÉïÌôîÎ©¥Ïóê ÏûàÎäî Ïù¥ÎØ∏ÏßÄÎ•º ÏÑ†ÌÉùÌï¥ÏÑú ÏóÖÎ°úÎìúÌïòÏÑ∏Ïöî.")
uploaded = files.upload()

# 3Ô∏è‚É£ ÏóÖÎ°úÎìúÌïú ÌååÏùº Ïù¥Î¶Ñ Í∞ÄÏ†∏Ïò§Í∏∞
filename = list(uploaded.keys())[0]
print(f"‚úÖ ÏóÖÎ°úÎìúÎêú ÌååÏùº: {filename}")

# 4Ô∏è‚É£ Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞ (Î∞îÏù¥Ìä∏ ‚Üí OpenCV Ïù¥ÎØ∏ÏßÄÎ°ú Î≥ÄÌôò)
file_bytes = np.frombuffer(uploaded[filename], np.uint8)
img = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

# 5Ô∏è‚É£ BGR ‚Üí RGB Î≥ÄÌôò
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 6Ô∏è‚É£ Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()

# 7Ô∏è‚É£ Í∞ÑÎã®Ìïú Ïù∏Ïãù ÏòàÏãú (Canny Edge Detection)
edges = cv2.Canny(img, 100, 200)
plt.figure(figsize=(6,6))
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('Edge Detection')
plt.show()
```
